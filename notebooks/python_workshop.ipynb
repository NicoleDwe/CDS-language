## Analyzing data of multiple files

```python
import os
import glob
import matplotlib.pyplot as plt
import numpy
```

```python
file_pattern = os.path.join("..", "data", "python-workshop", "inflammation*.csv")
filenames = sorted(glob.glob(file_pattern))
```

```python
filenames
```

```python
filenames_subset = filenames[:3]
filenames_subset
```

```python
for filename in filenames_subset:
    print(filename)
    data = np.loadtxt(fname = filename, delimiter = ",")
    # visualise
    fig = plt.figure(figsize = (10.0, 3.0))
    # three subplots
    axes1 = fig.add_subplot(1,3,1)
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)
    
    axes1.set_ylabel("average")
    meanval = np.mean(data, axis = 0)
    axes1.plot(meanval)
    
    axes2.set_ylabel("max")
    maxval = np.max(data, axis = 0)
    axes2.plot(maxval)
    
    axes3.set_ylabel("min")
    minval = np.min(data, axis = 0)
    axes3.plot(minval)
    
    fig.tight_layout()
    plt.show()
    #plt.savefig("directory")
    
```

## Clean up visualisation from function

```python
def visualize(filename):
    
    """ Visualization of mean, max and min values in tabular data at file
    Parameter:
        - filename: str variable, path to file/tabular data object
    """
    
    # import data
    data = np.loadtxt(fname = filename, delimiter = ",")
    
    # make canvas
    fig = plt.figure(figsize = (10.0, 3.0))
    
    # three subplots
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)
    
    # plot1: mean value over days
    axes1.set_ylabel("average")
    axes1.set_xlabel("days")
    axes1.plot(np.mean(data, axis = 0))
    
    # plot2: max value over days
    axes2.set_ylabel("max")
    axes2.set_xlabel("days")
    axes2.plot(np.max(data, axis = 0))
    
    # plot3: min value over days
    axes3.set_ylabel("min")
    axes3.set_xlabel("days")
    axes3.plot(np.min(data, axis = 0))
    
    # make it non-tight
    fig.tight_layout()
    plt.show()
```

```python
file_pattern = os.path.join("..", "data", "python-workshop", "inflammation*.csv")
filenames = sorted(glob.glob(file_pattern))
filenames_subset = filenames[:3]
```

```python
for filename in filenames: 
    visualize(filename)
```

## Testing suspicious behaviour

```python
def detect_problems(filename):
    """
    detect issues in data for maxima and minima
    
    input:
        - filenmae: str path to file
    """
    data = np.loadtxt(fname = filename, delimiter= ",")
    
    if np.max(data, axis = 0)[0] == 0 and np.max(data, axis = 0)[20] == 20:
        print("suspicious looking maxima!")
        
    elif np.sum(np.min(data, axis = 0)) == 0:
        print("suspicious looking minima")
        
    else:
        print("everything seems ok")
```

```python
filenames = sorted(glob.glob(os.path.join("..", "data", "python-workshop", "inflammation*.csv")))
for filename in filenames:
    print(filename)
    visualize(filename)
    detect_problems(filename)
```
